## Total no of attacks per year

```python
x_year = fin_df['Year'].unique()
y_count_years = fin_df['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (18,10))
sns.barplot(x = x_year,
           y = y_count_years,
           palette = 'rocket')
plt.xticks(rotation = 45)
plt.xlabel('Attack Year')
plt.ylabel('Number of Attacks each year')
plt.title('Attack_of_Years')
plt.show()
```

###  A decrease in the number of terror strikes since 2014,which means there is an increase in defense 

### NO OF ATTACKS FOR EACH COUNTRY

```python
plt.subplots(figsize=(15,6))
country_attacks = fin_df.Country.value_counts()[:16].reset_index()
country_attacks.columns = ['Country','Total Attacks']
sns.barplot(x=country_attacks.Country, y=country_attacks['Total Attacks'], palette= 'OrRd_r',edgecolor=sns.color_palette('dark',10))
plt.xticks(rotation=30)
plt.title('Number of Total Attacks in Each Country')
plt.show()

```

### The most number of terror attacks has taken place in Iraq

```python
region_year = pd.crosstab(fin_df.Year, fin_df.Region)

region_year.head(150)
```

```python
color_list_reg_yr = ['palegreen', 'lime', 'green', 'Aqua', 'skyblue', 'darkred', 'darkgray', 'tan', 'orangered', 'plum', 'salmon', 'mistyrose']
region_year.plot(figsize=(14, 10), fontsize=12, color=color_list_reg_yr)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Attacks in a Year', fontsize=12)
plt.title(' Region Based Terrorist Attacks over Years', fontsize=12)
plt.show()
```

```python
fin_df['Year10'] = pd.cut(fin_df['Year'], bins=[1969, 1980, 1990, 2000, 2010, 2020], 
               labels=['1970-1980', '1980-1990', '1990-2000', '2000-2010', '2010-2020'])
fin_df
```

```python
region_year10 = pd.crosstab(fin_df.Year10, fin_df.Region,)
region_year10
```

```python
color_list_reg_yr = ['palegreen', 'lime', 'green', 'Aqua', 'skyblue', 'darkred', 'darkgray', 'tan', 'orangered', 'plum', 'salmon', 'mistyrose']

region_year10.plot(figsize=(14, 10), fontsize=12, color=color_list_reg_yr)
plt.xlabel('Years', fontsize=12)
plt.ylabel('Number of Attacks in Each Decade', fontsize=12)
plt.title(' Region Based Terrorist Attack over Decades', fontsize=12)
plt.show()
```

### Terrorist activities went down in South & Central America considerably and there's a drastic rise in Middle East and South Asia.

```python
weapon_type_Region_year10 = pd.crosstab(fin_df.Year10,fin_df.Weapon_type)
weapon_type_Region_year10.head(6)
```

```python
color_list_weapon_yr = ['lime', 'palegreen', 'darkred', 'Aqua', 'orangered', 'salmon', 'tan', 'lavender', 'plum', 'darkgray', 'magenta', 'mistyrose']

weapon_type_Region_year10.plot(figsize=(14, 10), fontsize=12, color=color_list_weapon_yr)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Types of Weapon Used in Each Decade', fontsize=12)
plt.title(' Weapons Used over Decades by Terrorists', fontsize=12)
plt.show()
```
